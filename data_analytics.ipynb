```python
#load the data
data= pd.read_csv('data_spam.csv' , encoding='latin-1')
data
```

```python
data.columns
```

```python
# drop the anonymous cols
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'] , axis =  1 , inplace = True)
data
```

```python
#basic preprocessing drop the duplicate rows
data.drop_duplicates(inplace=True)
data.shape
```

```python
#check for missing values in cols
data.isnull().sum()
#no missing data
```

```python
#change cols name v1 to type and v2 to email
data = data.rename(columns={'v1': 'Type', 'v2': 'Email'})
data
```

```python
# convert the class to binary classification as sklearn deals with numerical data only
data.classify[data.classify == 'spam'] = 1
data.classify[data.classify == 'ham'] = 0
data
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer
import string

def process_mail (mail):
  #remove stopwords
  #remove punc
  # change to lower case
  # perform lemmatization
  stops = set(stopwords.words("english"))
  nonpunc = [char for char in mail if char not in string.punctuation]
  nonpunc = ''.join(nonpunc)
  nonpunc = nonpunc.lower()
  words = [word for word in nonpunc.split() if word not in  stops]
  lemmatizer=WordNetLemmatizer()
  clean_words = [lemmatizer.lemmatize(w) for w in words]
  return clean_words
```

```python
data['Email'].head().apply(process_mail)
```

```python
#filter tokens by Pos tags
nltk.download('averaged_perceptron_tagger')
data['splitted'] = data['Email'].str.split(" ")
data['splitted']
```

```python
data['tagged'] = data['splitted'].apply(nltk.pos_tag)
data['tagged']
```

```python
data.drop(['splitted'] , axis =  1 , inplace = True)
data
```

```python
#feature generation
# bag of words is one word embedding techinques
#we can determine which to use based on the accuracy
from sklearn.feature_extraction.text import CountVectorizer
mail = CountVectorizer(analyzer=process_mail).fit_transform(data['Email'])
print(mail)
```

```python
from sklearn.model_selection import train_test_split

#split the data into 80:20
x_train, x_test, y_train, y_test = train_test_split(mail, data['Email'], test_size = 0.2, random_state = 0)
print(x_train)
```

```python
# NAIVE BAYES CLASSIFIER
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.tree import DecisionTreeClassifier

# train
classifier = MultinomialNB().fit(x_train, y_train)

# evaluate the model
prediction = classifier.predict(x_test)

accuracy = accuracy_score(y_test, prediction)
recall = recall_score(y_test, prediction)
precision = precision_score(y_test, prediction)
f1_score1 = f1_score(y_test, prediction)

# printing
print('Accuracy: ',accuracy)
print('Recall: ',recall)
print('Precision: ',precision)
print('F1-score: ',f1_score1)
```

```python
# SVM
from sklearn import svm

#train

# when using kernel, accuracy = 0.97 otherwise when not using, accuracy = 0.95
model = svm.SVC(kernel = 'linear')
model.fit(x_train, y_train)

#evaluate
prediction_svm= model.predict(x_test)

accuracy_svm = accuracy_score(y_test,prediction_svm)
recall_svm = recall_score(y_test, prediction_svm)
precision_svm = precision_score(y_test, prediction_svm)
f1_score1_svm = f1_score(y_test, prediction_svm)

#printing
print('SVM Accuracy:',accuracy_svm)
print('recall: ',recall_svm)
print('Precision: ',precision_svm)
print('F1-score: ',f1_score1_svm)
```

```python
#Creating & Training Data With Decision Tree
Entropy = DecisionTreeClassifier(criterion = "gini" , random_state = 0 ,max_depth = 10 , min_samples_leaf=6)
Entropy.fit(x_train,y_train)

#Evaluting the data
pred = Entropy.predict(x_test)

# printing
print('Accuracy : ', accuracy_score(y_test,pred))
print('Precision : ', precision_score(y_test, pred))
print('Recall : ', recall_score(y_test, pred))
print('f1_score : ', f1_score(y_test,pred))
```
